# <font color="red">First Use the Gready Algorithm</font>

```python
def cashierGready(coins, sum):
    n = len(coins)
    numberOfCoins = 0
    for i in range(n-1, -1, -1):
        while(sum >= coins[i]):
            numberOfCoins = numberOfCoins + 1
            sum = sum - coins[i]
        if (sum == 0):
            break
    return numberOfCoins
```

```python
coins = coinSet1.copy()
```

```python
sum = 1211
cashierGready(coins, sum)
```

```python
sum = 35
cashierGready(coins, sum)
```
