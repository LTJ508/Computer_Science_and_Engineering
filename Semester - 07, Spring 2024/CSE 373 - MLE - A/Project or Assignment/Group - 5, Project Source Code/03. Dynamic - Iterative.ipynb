# <font color="red">Dynamic Programing - Iterative</font>

```python
#defining the algorithm
def cashierDynamicIterative(coins, sum):
    dp = [0] * (sum+1) #initialize the daynamic memorizing array
    dp[0] = 1 #initial result, there is only one way to make sum 0 by taking no coin
    
    for coin in coins: #this loop will run O(number of Coin)
        for i in range(coin, sum + 1): #this loop will run O(sum)
            dp[i] += dp[i - coin]
            
    return dp[sum]
```
