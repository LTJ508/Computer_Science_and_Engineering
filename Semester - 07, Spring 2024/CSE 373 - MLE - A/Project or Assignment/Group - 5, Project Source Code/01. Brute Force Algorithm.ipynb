# <font color="red">Brute Force Algorithm - All Possible Combinations</font>

```python
#defining the algorithm
def cashierBruteForceCombinations(coins, sum, startIndex):
    if (sum == 0): #base case 1
        return 1
    if (sum < 0 or startIndex >= len(coins)): # base case 2
        return 0
    with_startIndex = cashierBruteForceCombinations(coins, sum - coins[startIndex], startIndex) #calling recursively with the starting index
    without_startIndex = cashierBruteForceCombinations(coins, sum, startIndex+1) #calling recursively without the starting index
    return with_startIndex + without_startIndex
```

# <font color="red">Plotting Graph</font>

## In our problem we have two variable, one is number of coim, another is sum
Execution time depends on this two variable.\
So we first fixed a sum amount and check for multiple coin set, identify how its behave with the changing of number of coin in a coinset.\
Then we fixed a coin set and check for multiple sum amount, identify how this algorithm behave with the changing of sum amount.\
As our algorithm have two for loop, outer for loop will run O(number of coin) and inner for loop will run O(sum).\
Therefore, our time complexity will be O(number of coin * sum). Which is linear function.\
Lets verify with the graph

```python
#Random Sum value generated randomly by random library, and used in every algorithm for identify the time complexity
random_sum = [150, 270, 351, 370, 385, 431, 510, 675, 831, 987]
```

```python
#Random Coin Set generated randomly by random library, and used in every algorithm for identify the time complexity
CoinSet = [[37, 51, 73],
           [54, 73, 80, 93],
           [6, 17, 25, 33, 64],
           [6, 9, 11, 22, 29, 76],
           [25, 26, 33, 47, 78, 90, 99],
           [16, 28, 31, 64, 78, 83, 86, 96],
           [4, 37, 44, 48, 53, 65, 67, 86, 94],
           [12, 14, 21, 22, 29, 50, 57, 58, 74, 92],
           [5, 10, 11, 16, 56, 66, 77, 82, 90, 94, 96],
           [6, 10, 20, 23, 33, 38, 39, 41, 62, 87, 96, 98]]
```

```python
#variable that can adjust the plotting parameter
fixed_sum = 987
coins = CoinSet[9]
```

### Increasing the recursive depth limit

```python
# Set recursion limit
sys.setrecursionlimit(fixed_sum*10)

# Example recursive function
def recursive_function(n):
    if n <= 0:
        return 0
    return recursive_function(n - 1) + 1

# Test the function
try:
    print(recursive_function(fixed_sum))
except RecursionError as e:
    print("Recursion limit exceeded:", e)
    sys.setrecursionlimit(1000)  # back to a safer default
```

## This Cell
will generate the data point for ploting the graph.\
Here Sum amount is fixed and coinSet is variable

```python
#declaring two variable for storing the data points
times = []
num_coins = []

#this loop will change the coin set and call the function to calculate the execution time
for coins in CoinSet:
    start_time = time.time() #time count start
    numberOfCombinations = cashierBruteForceCombinations(coins, fixed_sum, 0)
    end_time = time.time() #time count end

    #storing the data point for ploting
    times.append(end_time - start_time)
    num_coins.append(len(coins))

    #printing the result info, how its changing
    print("\nNumber of combinations:", numberOfCombinations)
    print("Number of Coins:", len(coins))
    print("Sum:", fixed_sum)
    print("Execution time: {:.6f} seconds".format(end_time - start_time))
```

```python
#using the matplotlib library
plt.figure(figsize=(10, 6))
plt.plot(num_coins, times, 'o-')
plt.xlabel('Number of Coins in Coin Set')
plt.ylabel('Execution Time (seconds)')
plt.title(f'Execution Time vs. Number of Coins (Sum = {fixed_sum})')
plt.grid(True)
plt.show()
```

### Ploting the Polynomial Fit Graph

```python
coefficients_polynomial = np.polyfit(num_coins, times, 2) #for quadratic function, where degree is 2
polynomial = np.poly1d(coefficients_polynomial)

# Generate x values for plotting the polynomial
x_poly = np.linspace(min(num_coins), max(num_coins), 500)
y_poly = polynomial(x_poly)

plt.figure(figsize=(10, 6))
plt.plot(num_coins, times, 'o', label='Data Points')
plt.plot(x_poly, y_poly, '-', label=f'Polynomial Fit: {polynomial}')
plt.xlabel('Number of Coins in Coin Set')
plt.ylabel('Execution Time (seconds)')
plt.title(f'Execution Time vs. Number of Coins (Sum = {fixed_sum})')
plt.legend()
plt.grid(True)
plt.show()
```

### Ploting the Linear Fit Graph
As we can see the data point are rising as a linear graph

```python
coefficients_linear = np.polyfit(num_coins, times, 1)
linear = np.poly1d(coefficients_linear)

# Generate x values for plotting the Linear
x_linear = np.linspace(min(num_coins), max(num_coins), 500)
y_linear = linear(x_linear)

plt.figure(figsize=(10, 6))
plt.plot(num_coins, times, 'o', label='Data Points')
plt.plot(x_linear, y_linear, '-', label=f'Linear Fit: {linear}')
plt.xlabel('Number of Coins in Coin Set')
plt.ylabel('Execution Time (seconds)')
plt.title(f'Execution Time vs. Number of Coins (Sum = {fixed_sum})')
plt.legend()
plt.grid(True)
plt.show()
```

## This Cell

```python
#declaring two variable for storing the data points
times = []
sums = []

#this loop will change the coin set and call the function to calculate the execution time
for sum_value in random_sum:
    start_time = time.time() #time count start
    numberOfCombinations = cashierBruteForceCombinations(coins, sum_value, 0)
    end_time = time.time() #time count end

    #storing the data point for ploting
    times.append(end_time - start_time)
    sums.append(sum_value)

    #printing the result info, how its changing
    print("\nNumber of combinations:", numberOfCombinations)
    print("Number of Coins:", len(coins))
    print("Sum:", sum_value)
    print("Execution time: {:.6f} seconds".format(end_time - start_time))
```
