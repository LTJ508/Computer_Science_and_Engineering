# <font color="red">Gready Algorithm</font>

```python
def cashierGready(coins, sum):
    n = len(coins)
    numberOfCoins = 0
    for i in range(n-1, -1, -1):
        while(sum >= coins[i]):
            numberOfCoins = numberOfCoins + 1
            sum = sum - coins[i]
        if (sum == 0):
            break
    return numberOfCoins
```

# <font color="red">Brute Force Algorithm</font>

```python
numberOfCoins = float('inf')
```

```python
def cashierBruteForce(sum, count):
    global numberOfCoins
    global coins
    if (sum == 0):
        numberOfCoins = min(numberOfCoins, count)
        return
    for coin in coins:
        if (coin <= sum):
            cashierBruteForce(sum - coin, count + 1)
    return numberOfCoins
```

# <font color="red">Brute Force Algorithm - All Possible Permutations</font>

```python
numberOfWays = 0
```

```python
def cashierBruteForcePermutations(coins, sum):
    global numberOfWays
    if (sum == 0):
        numberOfWays = numberOfWays + 1
        return
    for coin in coins:
        if (coin <= sum):
            cashierBruteForcePermutations(coins, sum - coin)
    return numberOfWays
```
